```python
import pandas as pd
import numpy as np
```

```python
google = pd.read_csv('./results/new_google_normalized_trends.csv')
```

```python
google
```

```python
google['Average'] = google.iloc[:, 1:].sum(axis=1)/13
```

```python
google.sort_values(by=['Words'], inplace=True)
google.reset_index(drop=True, inplace=True)
```

```python
google = google.transpose()
```

```python
google.to_csv('./tp_google_trends_n.csv')
```

```python
cdc = pd.read_csv('./cdc.csv')
```
