```python
import pandas as pd
```

```python
trend1 = pd.read_csv('./deaths_1-5.csv')
trend2 = pd.read_csv('./deaths_6-9.csv')
trend3 = pd.read_csv('./deaths_10-13.csv')
trend1_coeff = 32.391813/5.438596
trend1 = trend1.iloc[:, 1:] * trend1_coeff
trend3_coeff = 32.391813/7.754386
trend3 = trend3.iloc[:, 1:] * trend3_coeff
trend2 = trend2.iloc[:, 1:]
```

```python
trend1.drop(['Cardiovascular disease: (Worldwide)'], axis=1, inplace=True)
```

```python
trend2.drop(['Cardiovascular disease: (Worldwide)'], axis=1, inplace=True)
```

```python
all_trend = pd.concat([trend1, trend2, trend3], axis=1)
```

```python
all_trend = all_trend.transpose()
```

```python
newGoogle = pd.DataFrame()
for i in range(0,13):
    newGoogle[str(i+2004)] = all_trend.iloc[:, (i*12):(i+1)*12].sum(axis=1)
```

```python
newGoogle
```

```python
newGoogle_n = newGoogle / newGoogle.sum(axis=0)
```

```python
newGoogle_n.to_csv('./tp_google_trends_n.csv')
```
